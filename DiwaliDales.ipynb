```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('DiwaliSalesData.csv', encoding='unicode_escape')
df.head()
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.Amount = df['Amount'].fillna(df['Amount'].mean())
```

```python
df.info()
```

```python
df.drop(columns = ['Status','unnamed1'],inplace=True)
```

```python
df['Amount'] = round(df['Amount'])
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
df.rename(columns = {'Product_Category' : 'Category'},inplace = True)
df.head()
```

```python
df.describe()
```

Exploratory Data Analysis 

Gender distribution in users

```python
ax = sns.countplot(x='Gender',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'])['Amount'].sum().sort_values(ascending=False)
sns.barplot(sales_gen, palette=['#0077b6','#936639'])
```

From above graph we can say that, there are more Females buyers than Males and also the amount spent by females is higher 

AGE -

```python
age_order = sorted(df['Age Group'].unique()) # to get the group in order on x-axis

ax = sns.countplot(x='Age Group',data=df,hue='Gender',order=age_order) # it is for division in a single bar -here we are doing Male - Female division i n a age group 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Age Group'])['Amount'].sum().sort_values(ascending=False)
sns.barplot(sales_gen)
```

From above analysis we got the insight that most of the buyers are in 26-35 age group of which most are females

State

```python
df.columns
```

```python
sales_state1 = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(15, 6))
ax = sns.barplot(x='State', y='Orders', data=sales_state1,hue='State')
#plt.xticks(rotation=45) # to avoid overlapping on x-axis
for p in ax.patches: # to  mark order on every bar
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')

plt.show()
```

```python
sales_state2 = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(15, 6))
ax = sns.barplot(x='State', y='Amount', data=sales_state2,hue='State')
#plt.xticks(rotation=45) # to avoid overlapping on x-axis
for p in ax.patches: # to  mark order on every bar
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')

plt.show()
```

From above graph we can say that mostt ofd the orders are from UP, Maharashtra and Karanataka while bihar and gujrat had least

Marital Status

```python
ax = sns.countplot(x='Marital_Status',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data3 = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount', data=sales_data3,hue='Gender')
```

We got the insight that --- Most of the buyers are married women and they also have high purchasing power

Occupation

```python
plt.figure(figsize=(15, 6))
ax = sns.countplot(x='Occupation',data=df,hue='Occupation')
plt.xticks(rotation=15)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(15, 6))
plt.xticks(rotation=15)
sales_data4 = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount', data=sales_data4,hue='Occupation')
```

From above graph we can see that most of the buyers are working in IT, healthcare and aviation

Product Category

```python
plt.figure(figsize=(25, 6))
ax = sns.countplot(x='Category',data=df,hue='Category')
plt.xticks(rotation=25)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(15, 6))
sales_data4 = df.groupby(['Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(7)
sns.barplot(x='Category',y='Amount', data=sales_data4,hue='Category')
```

From above Analysis we came to the conclusion that most of the sold products are from Food , clothing and Electronics

```python
plt.figure(figsize=(15, 6))
sales_data4 = df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Amount', data=sales_data4,hue='Product_ID')
```

Conclusion ---

Married women in the age group 26-35 from UP , Karnatak and maharashtra working in IT, Healthcare and Aviation are more likely to buy products from food , clothing and electronics category .
